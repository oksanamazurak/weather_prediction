# Exploratory Data Analysis (EDA) – WeatherAUS Dataset

This notebook performs a basic exploratory data analysis of the **WeatherAUS** dataset.

Steps included:
1. Import libraries
2. Load dataset
3. Inspect dataset structure
4. Check missing values
5. Visualize target variable distribution
6. Build correlation matrix


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df = pd.read_csv('../data/raw/WeatherAUS.csv')
df.head()
```

## Dataset Shape

```python
df.shape
```

## Dataset Information

```python
df.info()
```

## Missing Values

```python
df.isnull().sum()
```

## Target Variable Distribution

```python
sns.countplot(x='RainTomorrow', data=df)
plt.title('Distribution of RainTomorrow')
plt.show()
```

## Convert Yes/No to Numeric

```python
df['RainTomorrow'] = df['RainTomorrow'].map({'Yes':1, 'No':0})
df['RainToday'] = df['RainToday'].map({'Yes':1, 'No':0})
```

## Correlation Matrix

```python
numeric_df = df.select_dtypes(include=['float64','int64'])
corr = numeric_df.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr)
plt.title('Correlation Matrix')
plt.show()
```

## Temperature Distribution

```python
sns.histplot(df['Temp3pm'], kde=True)
plt.title('Temperature Distribution at 3pm')
plt.show()
```

## Humidity Distribution

```python
sns.histplot(df['Humidity3pm'], kde=True)
plt.title('Humidity Distribution at 3pm')
plt.show()
```
